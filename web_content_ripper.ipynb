```python
from bs4 import BeautifulSoup
import re

import lxml
from lxml.html.clean import Cleaner
from lxml.html import document_fromstring

cleaner = Cleaner()
cleaner.javascript = True # This is True because we want to activate the javascript filter
cleaner.style = True      # This is True because we want to activate the styles & stylesheet filter


EXTRA_SPACES = re.compile("^\s*")
EXTRA_LINES = re.compile("^\n")
ADS_WORDS = re.compile("(\(https://\)　　天才本站地址：。手机版阅读网址：)")

def remove_html_tags(text):
    
    text = document_fromstring(text)
    text = lxml.html.tostring(cleaner.clean_html(text))
    
    text = BeautifulSoup(text, "html.parser").get_text() # Remove HTML tags
    text = EXTRA_SPACES.sub("", text)
    text = EXTRA_LINES.sub("", text)
    text = ADS_WORDS.sub("", text)
    return text
```
